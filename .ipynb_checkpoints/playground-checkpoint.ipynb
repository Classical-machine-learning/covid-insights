# Creating a corpus from all the texts

```python
titles_corpus = []
abstract_corpus = []
text_corpus = []
```

```python
for pap in papers:
    temp = json.load(open(pap, "r"))
    titles_corpus.append(temp["metadata"]["title"])
    abstract_corpus.extend([a["text"] for a in temp["abstract"]])
    text_corpus.extend([a["text"] for a in temp["body_text"]])
```

```python
len(titles_corpus), len(papers)
```

```python
from nltk import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
from wordcloud import WordCloud
```

## Sent tokenize

```python
tit_tok = sent_tokenize(" ".join(titles_corpus))
```

```python
abs_tok = sent_tokenize(" ".join(abstract_corpus))
```

```python
txt_tok = sent_tokenize(" ".join(text_corpus))
```

## Word tokenize

```python
tit_tok_w = word_tokenize(" ".join(titles_corpus))
```

```python
abs_tok_w = word_tokenize(" ".join(abstract_corpus))
```

```python
txt_tok_w = word_tokenize(" ".join(text_corpus))
```

## Other preprocess steps

```python
tit_tok_w
```

```python
def pipeline(lis):
    lis = [a.lower() for a in lis]
    lis = [word for word in lis if word.isalnum()]
    lis = [w for w in lis if not w in stop_words]
    return lis
```

# Most common words

## Titles

```python
Counter(pipeline(txt_tok_w))
```

```python
wordcloud = WordCloud(
    width=1000, height=1000, background_color="white", min_font_size=10, max_words=2000
)
wordcloud.generate(" ".join(pipeline(tit_tok_w)))
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

## Abstracts

```python
wordcloud = WordCloud(
    width=1000, height=1000, background_color="white", min_font_size=10, max_words=2000
)
wordcloud.generate(" ".join(pipeline(abs_tok_w)))
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

# Creating a Text generation algo to answer potential queries

```python
from fastai.text import *
```

```python
data_frame_text = pd.DataFrame(txt_tok)
data_frame_text.head(4)
```

```python
data2 = (
    TextList.from_df(data_frame_text, cols=0)
    .split_by_rand_pct(0.1)
    .label_for_lm()
    .databunch(bs=16)
)
```

```python
data2
```

```python
learn = language_model_learner(data2, AWD_LSTM, drop_mult=0.3)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(1, 1e-4, moms=(0.8, 0.7))
```

```python
learn.unfreeze()
learn.fit_one_cycle(10, 1e-4, moms=(0.8, 0.7))
```

```python
learn.save("stage-1")
```

```python
print(learn.predict("confirmed cases in china", 50, temperature=0.8))
```
